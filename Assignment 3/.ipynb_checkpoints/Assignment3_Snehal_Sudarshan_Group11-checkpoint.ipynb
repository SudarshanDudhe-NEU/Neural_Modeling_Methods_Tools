### Wine Quality Prediction

This notebook focuses on predicting the quality of wine using a neural network model. The dataset utilized comprises various physicochemical properties of red wines, and the primary objective is to accurately predict wine quality based on these features. By leveraging machine learning techniques, we aim to uncover the relationships between the properties and the resulting quality ratings.


## Load the wine quality dataset

```python
wines = pd.read_csv('data/winequality-red.csv')
wines.head(20)
```

#### Feature Engineering

Feature engineering techniques are applied to enhance the dataset and potentially improve model performance. This may include creating new features from existing ones, scaling or normalizing features, and employing dimensionality reduction techniques.


## Separate features and target variable

```python
X_train = wines.drop('quality', axis=1).values.astype(float)
y_train = wines['quality']
```

```python
# Input variables
X_train
```

```python
# Target variable
y_train
```

#### Target Conversion to Categorical

For classification tasks, especially multiclass classification, the target variable must be in a format suitable for the model to learn effectively. One-hot encoding transforms the target into a binary matrix representation where each class is represented by a binary vector of 0s except for the class's index, which is marked as 1.


```python
from tensorflow.keras.utils import to_categorical

# Map quality scores to start from 0
y_mapped = y_train - y_train.min()

# Convert mapped quality scores to one-hot
y_categorical = to_categorical(y_mapped).astype(int)
y_categorical
```

```python
X_train.shape, y_categorical.shape
```

### Standard Scaler

The Standard Scaler standardizes features by removing the mean and scaling to unit variance, ensuring that the data is centered around zero with a standard deviation of one, which can enhance the convergence and performance of machine learning models.


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

## Let's Build the Neural Network Model

In this section, we will construct a neural network model using TensorFlow.


```python
import tensorflow as tf
tf.__version__
```

## NN Model 1

- **Architecture**: 1 hidden layer (10 neurons, ReLU)
- **Output Layer**: 6 neurons (Softmax)
- **Loss**: Mean Squared Error
- **Optimizer**: Adam
- **Metrics**: Accuracy


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(11))
model.add(Activation("relu"))
model.add(Dense(10))
model.add(Activation("softmax"))
model.add(Dense(6))
model.compile(loss='mean_squared_error', optimizer="adam", metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_categorical, epochs=1000, batch_size=30) 
```

```python
# Make predictions with the model
predictions_categorical = model.predict(X_train)

# Convert predictions from one-hot encoded format back to original label format
# The `argmax` method finds the index of the maximum value along axis 1 (across columns),
# which corresponds to the predicted class. add `y_train.min()` to map it back to the original range.
predictions = np.argmax(predictions_categorical, axis=1) + y_train.min()

# Convert `y_train` to a numpy array for easier comparison
y_train_array = np.array(y_train)

# Compare predictions with actual values
for i in range(10):
    print(f"Actual: {y_train_array[i]}, Predicted: {predictions[i]}")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy using scikit-learn's function
accuracy_sklearn = accuracy_score(y_train_array, predictions)
print(f"Accuracy (scikit-learn): {accuracy_sklearn:.2f}")

# classification report
print("\nClassification Report:")
print(classification_report(y_train_array, predictions))
```

```python
# y_train_array contains the actual labels and predictions contains the predicted labels from your model
accuracy_percentage = accuracy_score(y_train_array, predictions) * 100

print(f"Model Accuracy: {accuracy_percentage:.2f}%")
```

## Factors to take into consideration to Improve Model Accuracy

Achieving higher accuracy in a neural network involves tuning various components, including activation functions, hyperparameters, model architecture, and optimization strategies. Below are key areas to experiment with and considerations for each:

---

### 1. **Activation Functions**
The choice of activation function plays a significant role in how well the model learns and converges. Each function has specific use cases:

- **ReLU (Rectified Linear Unit)**:
  - **Usage**: Ideal for hidden layers in most deep learning architectures.
  - **Benefits**: Prevents vanishing gradient problems common in sigmoid and tanh. Computationally efficient.
  - **Drawbacks**: May suffer from the "dying ReLU" problem (neurons output zero constantly).
  - **Suggestion**: Use ReLU or variants like Leaky ReLU in hidden layers for faster convergence.

- **Leaky ReLU**:
  - **Usage**: Alternative to ReLU when you encounter the "dying ReLU" problem.
  - **Benefits**: Allows small negative outputs to prevent neurons from dying.
  - **Suggestion**: Switch to Leaky ReLU if ReLU underperforms.

- **Sigmoid**:
  - **Usage**: Best for binary classification tasks, especially in the output layer.
  - **Benefits**: Outputs values in [0, 1], ideal for probabilities.
  - **Drawbacks**: Susceptible to vanishing gradient problems in deep networks.
  - **Suggestion**: Use in the output layer only for binary classification.

- **Tanh (Hyperbolic Tangent)**:
  - **Usage**: Alternative to sigmoid, outputs in the range [-1, 1].
  - **Benefits**: Zero-centered, useful for data with negative and positive values.
  - **Drawbacks**: Still prone to vanishing gradients, though slightly better than sigmoid.
  - **Suggestion**: Consider tanh in hidden layers for specific use cases, but prefer ReLU.

- **Softmax**:
  - **Usage**: The go-to activation for the output layer in multi-class classification.
  - **Benefits**: Provides a probability distribution, summing to 1 across all output classes.
  - **Suggestion**: Use in the final layer for multi-class classification problems.

---

### 2. **Tuning Hyperparameters**

Hyperparameter tuning is essential for optimizing model performance. Here are the key parameters:

- **Learning Rate**:
  - **Effect**: Controls how fast the model updates weights. A learning rate too high might lead to missing the optimal solution, while too low may cause slow convergence.
  - **Suggestion**: Start with 0.001 (default for Adam) and try lower values (e.g., 0.0001) if the model struggles to converge.

- **Batch Size**:
  - **Effect**: Smaller batch sizes provide more frequent updates but can lead to noisy gradients, while larger batches offer smoother updates but require more resources.
  - **Suggestion**: Start with batch sizes of 32 or 64. Larger sizes (128, 256) can be tested, but ensure they fit in memory and don’t slow down training.

- **Optimizer**:
  - **Effect**: The optimizer dictates how model weights are updated during training.
    - **Adam**: Typically the best choice for most tasks. Adaptive learning rates simplify tuning.
    - **RMSprop**: Effective for noisy tasks or recurrent networks.
    - **SGD**: Can be effective for large datasets but often requires careful tuning of learning rates.
  - **Suggestion**: Start with Adam for its adaptability and speed of convergence. Experiment with RMSprop if noise in data is a concern.

---

### 3. **Model Architecture**

The architecture—number of layers and neurons—directly affects the complexity the model can capture.

- **Number of Hidden Layers**:
  - **Effect**: More layers enable capturing complex patterns, but too many can lead to overfitting.
  - **Suggestion**: Start with 2-3 hidden layers. Add more (up to 4-5) if the model is underfitting but be cautious of overfitting. Use **dropout** to mitigate overfitting risk.

- **Number of Neurons per Layer**:
  - **Effect**: Determines the capacity of the model to capture data patterns. Too few neurons can underfit, while too many can lead to overfitting.
  - **Suggestion**: Begin with 64-128 neurons in the first hidden layer, then gradually reduce in subsequent layers (e.g., 128 → 64 → 32). Adjust based on performance.

- **Dropout**:
  - **Effect**: Randomly deactivates a portion of neurons during training, preventing overfitting.
  - **Suggestion**: Add dropout layers between hidden layers, with dropout rates of 0.2 to 0.5. Experiment to find the right balance between regularization and convergence.

---

### 4. **Number of Epochs and Early Stopping**

- **Effect**: The number of epochs controls how many times the model sees the entire dataset. Too few epochs may lead to underfitting, while too many could overfit.
- **Suggestion**: Use **early stopping** to monitor validation loss and stop training when improvements cease. Start with 1000 epochs but allow early stopping to halt training when the model plateaus.

---

### 5. **Data Preprocessing**

Preprocessing is critical for improving model accuracy and convergence.

- **Feature Scaling**:
  - **Effect**: Neural networks require normalized input data for effective weight updates and convergence.
  - **Suggestion**: Use **StandardScaler** (zero mean, unit variance) to standardize features. This is especially important for features with varying scales.

---

### 6. **Model Evaluation and Metrics**

Accuracy is not always the best indicator of model performance. Other metrics provide a more complete picture:

- **Confusion Matrix & F1-Score**:
  - **Effect**: Provides insight into model performance, particularly for imbalanced datasets. Accuracy might be misleading if one class dominates.
  - **Suggestion**: Plot a confusion matrix and compute **precision**, **recall**, and **F1-score** to assess the model’s true strengths and weaknesses. 

  ---
  
By focusing on these steps and experimenting with key elements like activation functions, learning rates, and architecture, you can significantly improve the accuracy of your neural network. Always track your experiments and document how each change affects model performance.

## Experimentation Plan

To efficiently improve model accuracy while saving time, we’ll focus on specific strategies that are most likely to have an impact, based on the nature of our problem (wine quality classification), and avoid unnecessary experimentation:

1. **Activation Functions**:
   - **Best fit**: Since our task involves classification, we should focus on **ReLU** (for hidden layers) and **sigmoid** (for binary classification) or **softmax** (for multi-class classification) in the output layer.
   - **Rationale**: ReLU is computationally efficient and avoids vanishing gradients, while sigmoid/softmax provides well-suited output for classification.
   - **Eliminate**: Tanh and pure linear activations are not necessary here since they’re slower or more prone to vanishing gradients.

2. **Number of Hidden Layers**:
   - **Best fit**: Start with 2-3 hidden layers and experiment up to 4 if necessary. Deep networks may not add much value here due to the moderate complexity of the wine dataset.
   - **Rationale**: Over-complicating the architecture with too many hidden layers can lead to overfitting and unnecessary computational cost.
   - **Eliminate**: More than 4 hidden layers would likely not be useful unless the dataset was much more complex.

3. **Neurons per Layer**:
   - **Best fit**: Use a decreasing number of neurons per layer (e.g., 32 → 16 → 8). Higher neuron counts at the beginning allow the model to capture more features, while fewer neurons in later layers allow for fine-tuning.
   - **Rationale**: This structure encourages the model to learn more complex representations in the earlier layers and refine them later, without over-complicating the model.
   - **Eliminate**: Avoid excessively large numbers of neurons (e.g., 100+) since the model might overfit or become computationally expensive.

4. **Learning Rate**:
   - **Best fit**: Start with a learning rate of 0.001 (default for Adam) and slightly decrease to 0.0001 if necessary.
   - **Rationale**: Too high a learning rate can cause the model to skip over optimal solutions, while too low can lead to slow convergence.
   - **Eliminate**: Learning rates above 0.01 should be avoided as they often cause instability in training.

5. **Optimizer**:
   - **Best fit**: Use **Adam** as the default optimizer, as it combines the benefits of adaptive learning rates with momentum. It works well for most neural networks, including classification tasks.
   - **Rationale**: Adam is widely used and effective for fast convergence, especially for moderately complex datasets like ours.
   - **Eliminate**: We can avoid experimenting with **SGD** initially, as it typically requires careful tuning of learning rates and momentum, leading to longer experimentation cycles.

6. **Number of Epochs and Early Stopping**:
   - **Best fit**: Start with 1000 epochs and enable **early stopping** to avoid overfitting.
   - **Rationale**: Early stopping will allow us to use a high epoch count without overfitting. It will stop training when validation performance stops improving.
   - **Eliminate**: Avoid setting too low an epoch count (e.g., < 200), as this may not allow the model enough time to converge properly.

7. **Batch Size**:
   - **Best fit**: Use a moderate batch size (e.g., 32 or 64).
   - **Rationale**: This balances training stability and speed. A too-small batch size might introduce noisy gradients, while a too-large batch size may slow down convergence.
   - **Eliminate**: Very large batch sizes (e.g., 512+) can be avoided unless the dataset is extremely large, which is not the case here.

---

By focusing on these best-fit options, we can significantly improve accuracy while avoiding time-consuming experiments that are less likely to yield useful results. Document each change carefully and assess its impact on both training and testing accuracy.
